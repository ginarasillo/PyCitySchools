```python
import pandas as pd
from IPython.core.display import display,HTML
```

```python
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
school_data.head()
```

```python
student_data.head()
```

```python
total_schools = school_data["school_name"].count()
total_schools
```

```python
total_students = student_data["student_name"].count()
total_students
```

```python
total_budget = school_data["budget"].sum()
total_budget
```

```python
average_math = student_data["math_score"].mean()
average_math
```

```python
average_reading = student_data["reading_score"].mean()
average_reading
```

```python
passing_math= student_data.query('math_score >=70')["Student ID"].count()/ total_students*100
passing_math
```

```python
passing_reading= student_data.query('reading_score >=70')["Student ID"].count()/ total_students*100
passing_reading
```

```python
overall_passing= student_data.query('reading_score >=70 and math_score >=70')["Student ID"].count()/ total_students*100
overall_passing
```

```python
total_df = pd.DataFrame({
    "Total schools":[total_schools],
    "Total students":[f"{total_students:,.2f}"],
    "Total budget":[f"${total_budget:,.2f}"],
    "Average Math Score":[f"{average_math:.2f}"],
    "Average Reading Score":[f"{average_reading:.2f}"],
    "% Passing Math":[f"{passing_math:.2f}%"],
    "% Passing Reading":[f"{passing_reading:.2f}%"],
    "% Overall Passing":[f"{overall_passing:.2f}%"]
    
})
display(HTML("<h2>District Summary</h2>"))
total_df
```

```python
#a=pd.DataFrame(student_data.groupby("school_name")[["math_score","reading_score"]].mean())
#b=pd.DataFrame(student_data.groupby("school_name").size())
#a.merge(b, left_index=True, right_index=True)
result= {}

for school in student_data.school_name.unique():
    school_students= student_data.query(f'school_name=="{school}"')
    total_students= len(school_students)
    average_reading = school_students["reading_score"].mean()
    average_math = school_students["math_score"].mean()
    passing_math= school_students.query('math_score >=70')["Student ID"].count()/ total_students*100

    passing_reading= school_students.query('reading_score >=70')["Student ID"].count()/ total_students*100
    
    overall_passing= school_students.query('reading_score >=70 and math_score >=70')["Student ID"].count()/ total_students*100
    result[school]={
        "Total students":f"{total_students}",
        "Average Math Score":f"{average_math:.2f}",
        "Average Reading Score":f"{average_reading:.2f}",
        "TStudents": total_students,
        "AMathScore": average_math,
        "AReadingScore": average_reading,
        "%PassingMath":passing_math,
        "%PassingReading": passing_reading,
        "%OverallPassing": overall_passing,
        "% Passing Math":f"{passing_math:.2f}%",
        "% Passing Reading":f"{passing_reading:.2f}%",
        "% Overall Passing":f"{overall_passing:.2f}%"
    }
    
result=pd.DataFrame(result).transpose()
```

```python
formatted_columns = ["Total students", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]
result[formatted_columns]
```

```python
school_index=school_data.set_index("school_name")
school_index
```

```python
merge_df= result.merge(school_index, left_index=True, right_index=True)
merge_df["Per Student Budget"]=merge_df.budget/merge_df["size"]
merge_df=merge_df.drop(columns=["School ID","size"])
merge_df[formatted_columns + ["type","budget","Per Student Budget"]]
```

```python
merge_df= merge_df.rename(columns={"type":"School Type",
                                "budget":"Total Budget",
                                    })
                                            
                                              
merge_df                                              
```

```python
school_summary=merge_df[["School Type","Total students","Total Budget","Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]]
school_summary['Total Budget']=school_summary['Total Budget'].map('${:,.2f}'.format)
school_summary['Per Student Budget']=school_summary['Per Student Budget'].map('${:,.2f}'.format)

display(HTML("<h2>School Summary</h2>"))
school_summary.sort_index()
```

```python
school_summary= school_summary.sort_values("% Overall Passing", ascending=False)

display(HTML("<h2>Top Performing Schools (By % Overall Passing)</h2>"))
school_summary.head()
                
```

```python
school_summary= school_summary.sort_values("% Overall Passing", ascending=True)
display(HTML("<h2>Bottom Performing Schools (By % Overall Passing)</h2>"))
school_summary.head()
```

```python
merge_df_2 = merge_df.copy()

merge_df_2["AMathScore"] = merge_df_2["AMathScore"].astype(float)
merge_df_2["AReadingScore"] = merge_df_2["AReadingScore"].astype(float)
merge_df_2["%PassingMath"] = merge_df_2["%PassingMath"].astype(float)
merge_df_2["%PassingReading"] = merge_df_2["%PassingReading"].astype(float)
merge_df_2["%OverallPassing"] = merge_df_2["%OverallPassing"].astype(float)

psg=merge_df_2["Per Student Budget"]  
bins=[0,584,630,645,675]
bin_names = ["< $584", "$585-629","$630-644","$645-675"]
merge_df_2["Spending ranges (Per Student)"] = pd.cut(psg, bins, labels=bin_names)

sch_spn_math_scores = merge_df_2.groupby(["Spending ranges (Per Student)"])["AMathScore"].mean()
sch_spn_reading_scores = merge_df_2.groupby(["Spending ranges (Per Student)"])["AReadingScore"].mean()
sch_spn_passing_math = merge_df_2.groupby(["Spending ranges (Per Student)"]).mean()["%PassingMath"]
sch_spn_passing_reading = merge_df_2.groupby(["Spending ranges (Per Student)"]).mean()["%PassingReading"]
sch_spn_overall_passing_rate = merge_df_2.groupby(["Spending ranges (Per Student)"]).mean()["%OverallPassing"]

score_by_school_spn_df = pd.DataFrame({"Average Math Score":sch_spn_math_scores,
                                    "Average Reading Score":sch_spn_reading_scores,
                                    "% Passing Math":sch_spn_passing_math,
                                    "% Passing Reading":sch_spn_passing_reading,
                                    "% Overall Passing Rate": sch_spn_overall_passing_rate})

display(HTML("<h2>Scores by School Spending</h2>"))
score_by_school_spn_df
```

```python



totst=merge_df_2["TStudents"].astype(float)  
bins=[0,1000,2001,5001]
bin_names = ["< 1000", "$1000-2000","$2000-5000"]
merge_df_2["School Size"] = pd.cut(totst, bins, labels=bin_names)
    
sch_spn_math_scores = merge_df_2.groupby(["School Size"])["AMathScore"].mean()
sch_spn_reading_scores = merge_df_2.groupby(["School Size"])["AReadingScore"].mean()
sch_spn_passing_math = merge_df_2.groupby(["School Size"]).mean()["%PassingMath"]
sch_spn_passing_reading = merge_df_2.groupby(["School Size"]).mean()["%PassingReading"]
sch_spn_overall_passing_rate = merge_df_2.groupby(["School Size"]).mean()["%OverallPassing"]

score_by_school_size_df = pd.DataFrame({"Average Math Score":sch_spn_math_scores,
                                    "Average Reading Score":sch_spn_reading_scores,
                                    "% Passing Math":sch_spn_passing_math,
                                    "% Passing Reading":sch_spn_passing_reading,
                                    "% Overall Passing Rate": sch_spn_overall_passing_rate})

display(HTML("<h2>Scores by School Size"))
score_by_school_size_df
```
